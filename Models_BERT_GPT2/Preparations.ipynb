```python

# mount drive
from google.colab import drive
drive.mount('/content/gdrive')

# set path to datasets
path = "/content/gdrive/My Drive/Models_BERT_GPT2/"
print(path)

# import pandas and numpy
import pandas as pd
import numpy as np

# import regex to remove the \n \t \r
import re
regex = re.compile(r'[\n\r\t]')

# function prepares and returns line for datasets
def prepareLine1(line):
  line = line.replace('if name == "','')
  line = line.replace('":',' : ')
  line = line.replace('":',' : ')
  line = line.replace(' message = "','')
  line = line.replace('"',' : ')
  line = line.replace(';',' ')
  line = line.replace('do you have any of these ingredients?','')
  line = line.replace('instead you can use','')

  return line
```
