```python
# Step 3: import logging, QuestionAnsweringModel and QuestionAnsweringArgs
import logging
from simpletransformers.question_answering import QuestionAnsweringModel
from simpletransformers.question_answering import QuestionAnsweringArgs
```

```python
# Step 4: model type and model name
model_type = "bert"
model_name = "bert-base-cased"
```

```python
# Step 5: training arguments
training_args = {
  "reprocess_input_data" : True,
  "overwrite_output_dir" : True,
  "use_cached_eval_features" : True,
  "output_dir" : f"/outputs/{model_type}",
  "best_model_dir" : f"/outputs/{model_type}/best_model",
  "max_seq_length" : 128,
  #"num_train_epochs" : 100,
  "num_train_epochs" : 5,
  "evaluate_during_training_steps" : 1000,
  "save_model_every_epoch" : False,
  "save_eval_checkpoints" : False,
  "n_best_size" : 3,
  "train_batch_size" : 128,
}
```

```python
# Step 6: download model
# You must create account on huggingface.co and create there a HF_TOKEN token in settings
# Then you must set this token in Google Colab
model = QuestionAnsweringModel(model_type, model_name, args = training_args)
```

```python
# import regex to remove the \n \t \r
import re
regex = re.compile(r'[\n\r\t]')
```

```python
# Alternative Products dataset

# alternative products list
alternativeProductsTrainingList = []

# variable
tempText = ""

# read file
file = open(path+"AlternativeProductsCleaned.txt", "r")
line = file.read()

# add each line
tempText += line

# get tokens
tokens = tempText.split(':')

# product and answer
product = ""
answer = ""

# index used when looping through tokens
index = 0

# index used when appending data to the list
rowIndex = 0

# loop thorugh tokens
for token in tokens:

  # line is contains keyword, product to be replaced
  if index%2==0:
    token = token.replace("'","")
    token = regex.sub(' ', token)
    product = token.strip()
    question = "What can be used instead of "+product+"?"
    # print(question)

  # line contains alternative products
  elif index%2!=0:
    alternatives = regex.sub(' ', token)
    alternatives = alternatives.strip()
    context = " you can use "
    answer_start = len(context) + 1
    context = context + alternatives + " if you do not have "+product
    answer = alternatives
    # print(answer_start)
    # print(context)
    # print(answer)

  # add question and text to the dataset
  if index>0 and index%2!=0 and rowIndex<60: # and datasetIndex<40
    id = ""

    if rowIndex<10:
      id="0000"+str(rowIndex)
    elif rowIndex<100:
      id="000"+str(rowIndex)
    elif rowIndex<1000:
      id="00"+str(rowIndex)

    row = {"context":context , "qas":[{"id":id, "question":question, "answers": [{"text":answer, "answer_start":answer_start}], "is_impossible":False}]}
    # print(row)

    alternativeProductsTrainingList.append(row)
    rowIndex = rowIndex + 1

  # increase index
  index = index+ 1
```

```python
# Utensils Description dataset

# utensils description list
utensilsDescriptinsTrainingList = []

# variable
tempText = ""

# read file
file = open(path+"UtensilsDescriptionCleaned.txt", "r")
line = file.read()

# add each line
tempText += line

# get tokens
tokens = tempText.split(':')

# keyword and answer
keyword = ""
answer = ""

# index used when looping through tokens
index = 0

# index used when appending data to the list
rowIndex = 0

# loop thorugh tokens
for token in tokens:

  # line is contains keyword, product to be replaced
  if index%2==0:
    token = token.replace("'","")
    token = regex.sub(' ', token)
    keyword = token.strip()
    question = "What is  "+keyword+"?"
    # print(question)

  # line contains alternative products
  elif index%2!=0:
    answer = regex.sub(' ', token)
    answer = answer.strip()
    answer_start = len(keyword) + 4
    context = keyword+" is "+answer +"."
    # print(answer_start)
    # print(context)
    # print(answer)

  # add question and text to the dataset
  if index>0 and index%2!=0 and rowIndex<60: # and datasetIndex<40
    id = ""

    if rowIndex<10:
      id="0000"+str(rowIndex)
    elif rowIndex<100:
      id="000"+str(rowIndex)
    elif rowIndex<1000:
      id="00"+str(rowIndex)

    row = {"context":context , "qas":[{"id":id, "question":question, "answer":answer, "answers": [{"text":answer, "answer":answer,"answer_start":answer_start}], "is_impossible":False}]}
    # print(row)

    utensilsDescriptinsTrainingList.append(row)
    rowIndex = rowIndex + 1

  # increase index
  index = index+ 1
```
